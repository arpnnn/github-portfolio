```python
#create a working copy of raw data
df_clean = df.copy() 
```

```python
df_clean.columns = (df_clean.columns
                    .str.strip()
                    .str.lower()
                    .str.replace(' ', '_')
                    .str.replace('/', '_')
                    .str.replace('(', ' ', regex=False)
                    .str.replace(')', ' ', regex=False)
                   )
```

```python
df_clean.columns
```

```python
# Rename columns to follow SQL naming conventions
df_clean.rename(columns={
    'development_site_square_feet': 'development_site_sqft',
    'type_of_dwelling_unit': 'dwelling_unit_type',
    'type_of_dwelling_unit_per_code': 'dwelling_unit_type_code',
    'condo_parcel_or_plat_ y_n ': 'condo_parcel_or_plat',
    'zoning_report_zone': 'zoning_reported_zone',
    'zoning_class': 'zoning_classification',
    'issued_date': 'permit_issued_date',
    'final_date': 'permit_final_date',
    'expiration_date': 'permit_expiration_date'
}, inplace=True)
```

```python
df_clean.columns
```

```python
# Drop redundant, repetitive, and system-generated columns that are not required for analysis
cols_to_drop = ['objectid', 'sort', 'status_color', 'villnumb', 'maf_id', 'devsite_id', 'recordnumber', 'parent_construction_permit', 
                'xcoord_stateplane', 'ycoord_stateplane', 'x', 'y']
df_clean.drop(columns= cols_to_drop, inplace = True)
```

```python
df_clean.columns #new columns
```

```python
#change data types
df_clean['year_finaled'] = df_clean['year_finaled'].astype('Int64')
df_clean['quarter_final'] = df_clean['quarter_final'].astype('Int64')
df_clean['council_district'] = df_clean['council_district'].astype('Int64')
```

```python
df_clean.info()
```

```python
#Convert census ID columns to string type
census_cols = ['geoid10', 'geoid20', 'census_block_2010', 'census_block_group_2010', 
               'census_block_2010', 'census_block_group_2010']
for cols in census_cols:
    df_clean[cols] = df_clean[cols].astype("string")
```

```python
df_clean.info()
```

```python
#drop redundant colmns
#quarter_and_year_issued: combines the year and quarter of the permit issue date
#quarter_and_year_final: same, but for the final date, e.g., "2023_2" 
df_clean.drop(columns=["quarter_and_year_issued", "quarter_and_year_final"], inplace=True)
```

```python
#description_of_work — redundant text, info already captured in structured columns.
df_clean.drop(columns=["description_of_work"], inplace=True)
```

```python
#make the date consistent
date_cols = ['application_date','permit_issued_date','permit_final_date']
for cols in date_cols:
    df_clean[cols] = pd.to_datetime(df_clean[cols], errors = 'coerce')
```

```python
#drop redundant year issued and year finaled
df_clean.drop(columns=['year_issued', 'year_finaled'], inplace=True)
```

```python
df_clean.info()
```

```python
#Validate Housing Unit Logic
df_clean["unit_mismatch"] = (
    df_clean["net_units_permitted"] != (
        df_clean["new_units_permitted"] - df_clean["demolished_units_permitted"])
)
```

```python
#Check if numbers are consistent
df_clean["unit_mismatch"] 
```

```python
#Check the null values
df_clean.isnull().sum().sort_values(ascending=False)
```

```python
#drop columns with more than 1000 missing values
cols_to_drop= [
    "master_user_permit_number", "most_recent_inspection_date",
     "most_recent_inspection_type", "permit_expiration_date"]
df_clean.drop(columns=cols_to_drop, inplace=True)
```

```python
df_clean.info()
```

```python
#keep only permits issued in 2012 or later
df_clean = df_clean[df_clean["permit_issued_date"].dt.year>= 2012]
```

```python
#check if the permit number is unique
df_clean['permit_number'].is_unique 
```

```python
#Ensure permit numbers are unique and remove duplicate records and keep the last
df_clean = df_clean.drop_duplicates(subset="permit_number", keep="last") 
```

```python
df_clean.columns
```

```python
#Redefine order of columns
final_cols = [
    #permit identifiers
    'permit_number',
    'permit_stage',
    'type_of_permit',

    #location info
    'project_address',
    'neighborhood',
    'community_reporting_area',
    'community_reporting_area_number',
    'council_district',
    'urban_village_name',
    'urban_village_type',
    'comp_plan_name', 
    'comp_plan_type',
    'longitude',
    'latitude',

    #zoning info
    'zoning_primary_zone',
    'zoning_category',
    'zoning_classification',
    'zoning_all_zones',
    'zoning_reported_zone',

    #units info
    'new_units_permitted',
    'demolished_units_permitted',
    'net_units_permitted',
    'sleeping_rooms_permitted',
    'dwelling_unit_type',
    'dwelling_unit_type_code',
    'condo_parcel_or_plat',
    'unit_mismatch',

    #Dates
    'application_date',
    'permit_issued_date',
    'permit_final_date',
    'quarter_issued',
    'quarter_final',

    #Location codes
    'census_block_2010', 
    'census_block_group_2010', 
    'geoid10',
    'census_block_2020', 
    'census_block_group_2020',
    'geoid20',

    #source
    'source',
    'development_site_sqft'
]

df_clean = df_clean[final_cols]
```

```python
# Export the cleaned dataset to an Excel file for further use
df_clean.to_excel("Cleaned_Residential_Building_Permits.xlsx", index=False)
```

```python
df = pd.read_excel("Cleaned_Residential_Building_Permits2.xlsx")
```

```python
#Get the count of permits issued in 2022 (by month)
df_2022 = df[df['permit_issued_date'].dt.year == 2022].copy()  # Using .copy() to avoid modifying the original dataframe
```

```python
# Create a 'Year-Month' column for grouping by month (using .loc[] to modify safely)
df_2022.loc[:, 'Year-Month'] = df_2022['permit_issued_date'].dt.to_period('M')
```

```python
# Group by 'Year-Month' and count the permits
permit_count_by_month = df_2022.groupby('Year-Month').size().reset_index(name='permit_count')
```

```python
# Save the result to an Excel file
permit_count_by_month.to_excel("permit_count_by_month_2022.xlsx", index=False)
```
