##### Reshaping Median Home Sale Price Data (Wide to Long Format)


```python
import pandas as pd
```

```python
# Load raw median home price data from the source file
df = pd.read_excel("med_sale_price.xlsx")
```

```python
df
```

```python
# Reshape the dataset from wide to long format to support visualization
df_long = df.melt(id_vars=["Region"], var_name = "Date", value_name ="Price $")
```

```python
# Convert 'Date' column to datetime format
df_long['Date'] = pd.to_datetime(df_long['Date'], errors='coerce')
```

```python
# Extract the year from the 'Date' column and create a new column 'Year'
df_long['Year'] = df_long['Date'].dt.year
```

```python
#Export the data to an excel file
df_long.to_excel("longformat_median_saleprice_with_year.xlsx", index=False)
```
